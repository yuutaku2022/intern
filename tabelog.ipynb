```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from time import sleep

service = Service('chromedriver.exe')
browser = webdriver.Chrome(service=service)
url = 'https://tabelog.com/kumamoto/'
browser.get(url)
sleep(1)

# HREFのリストを保持
HREF_LIST = []
i = 0

# HREF取得処理
while True:
    HREFS = browser.find_elements(By.CSS_SELECTOR, "a.list-rst__rst-name-target.cpy-rst-name")
    for HREF in HREFS:
        HREF_TITLE = HREF.get_attribute("href")
        HREF_LIST.append(HREF_TITLE)
    #print("[INFO] HREF_LIST:", HREF_LIST)
    try:
        browser.find_element(By.CSS_SELECTOR, "a.c-pagination__arrow.c-pagination__arrow--next").click()
        i += 1
        if i == 5:
            break
    except:
        break

# 各URLの詳細を取得
for EACH_LIST in HREF_LIST:
    browser.get(EACH_LIST)
    sleep(1)
    # タイトル取得
    title = browser.find_element(By.XPATH, '//h2/span').text
    print("[INFO] title:", title)
    
    # 評価取得
    try:
        review = browser.find_element(By.CSS_SELECTOR, ".rdheader-rating__score-val-dtl").text
        print("[INFO] review:", review)
    except Exception as e:
        print("[ERROR] 評価の取得に失敗:", e)

    # テーブル取得
    tables = browser.find_elements(By.CSS_SELECTOR, ".c-table.c-table--form")
    for table in tables:
        print("[INFO] table:", table.text)
# ブラウザを閉じる
browser.quit()
```
